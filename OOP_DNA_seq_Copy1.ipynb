```python
# A Python class for handling biological sequences
class Sequence:
    def __init__(self,name,sequence):
        self.name = name
        self.sequence = sequence
        self.residues = {}
    def search(self,pattern):
        return self.sequence.find(pattern)
    def molecularWeight(self):
        mwt = 0.0
        for residue in self.sequence:
            mwt += self.residues[residue]
        return mwt
    def validSequence(self):
        for residue in self.sequence:
            if not residue in self.residues:
                return False
        return True
```

```python
def compareNames(self,other):
    if self.name == other.name:
        return True
    else:
        return False
```

```python
# Define two sequences to be aligned
X = seq1sub
Y = seq2sub
# Get a list of the global alignments between the two sequences ACGGGT and ACG
# No parameters. Identical characters have score of 1, else 0.
# No gap penalties.
alignments = pairwise2.align.globalxx(X, Y)

# Use format_alignment method to format the alignments in the list
for a in alignments:
    print(format_alignment(*a))
```

```python
# Define two sequences to be aligned
X2 = seq1sub2
Y2 = seq2sub2
# Get a list of the global alignments between the two sequences ACGGGT and ACG
# No parameters. Identical characters have score of 1, else 0.
# No gap penalties.
alignments2 = pairwise2.align.globalxx(X2, Y2)

# Use format_alignment method to format the alignments in the list
for a in alignments2:
    print(format_alignment(*a))
```
